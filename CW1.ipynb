```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import statsmodels.api as sm
```

## <span style="color: blue;">1. read data</span>

```python
data = pd.read_csv('data.csv')
mpr = pd.read_excel('mpr.xls')
```

## <span style="color: blue;">2. data preview</span>

```python
data.head(n=10)
mpr.head()
```

## <span style="color: blue;">3. compute excess returns</span>

```python
stocks = data.columns[5:]
for i in stocks:
    data[i] = data[i]-data['Risk-Free']
```

## <span style="color: blue;">4. LR by OLS estimate</span>
### <span style="color: blue;">4.1 time series regression</span>

```python
result = pd.DataFrame(columns = ['coefs','t_statistics','p_value','portfolios'])
for i in data.columns[5:]:
    y = data[i]
    x = data.iloc[:,1:4]
    x1 = sm.add_constant(x)
    lr = sm.OLS(y,x1)
    model = lr.fit()
    res = pd.concat([model.params,model.tvalues,model.pvalues],axis=1)
    res.columns =  ['coefs','t_statistics','p_value']
    res['portfolios'] = i
    result = pd.concat([result,res],axis = 0)
result
```

```python
result[result['p_value']>0.05]
```

### <span style="color: blue;">4.2 cross-section regression</span>

```python
y = mpr.iloc[:,1:].mean(axis=0)
HML = result[result.index=='HML']['coefs']
SMB = result[result.index=='SMB']['coefs']
MKT = result[result.index=='Mkt (vwretd)']['coefs']
x = pd.DataFrame({'MKT':MKT.values,'SMB':SMB.values,'HML':HML.values})
x.index = y.index
x1 = sm.add_constant(x)
lr2 = sm.OLS(y,x1)
model2 = lr2.fit()
model2.summary()
```

## <span style="color: blue;">5. CPAM fit (extended)</span>

```python
result = pd.DataFrame(columns = ['coefs','t_statistics','p_value','stocks'])
sigma2 = []
for i in data.columns[5:]:
    y = data[i]
    x = data.iloc[:,1]
    x1 = sm.add_constant(x)
    lr = sm.OLS(y,x1)
    model = lr.fit()
    res = pd.concat([model.params,model.tvalues,model.pvalues],axis=1)
    res.columns =  ['coefs','t_statistics','p_value']
    res['stocks'] = i
    result = pd.concat([result,res],axis = 0)
    sigma2.append(model.scale)
result
sigma2
```

```python
y = mpr.iloc[:,1:].mean(axis=0)
MKT = result[result.index=='Mkt (vwretd)']['coefs']
x = pd.DataFrame({'MKT':MKT.values,'sigma2':sigma2})
x.index = y.index
x1 = sm.add_constant(x)
lr2 = sm.OLS(y,x1)
model2 = lr2.fit()
model2.summary()
```
